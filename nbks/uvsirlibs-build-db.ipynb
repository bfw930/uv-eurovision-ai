```python
# data array processing
import numpy as np

# import ipython display, ipd.Audio playback widget
import IPython.display as ipd
```

```python
''' updated orchestration process '''

# define directory to search
base_path = '../data/audio/'

# define properties
props = {
    'file_ext': 'mp3',
    'file_limit': 100,
}

# build segment database from tracks within directory
seg_db = uvsirlibs.new_core.build_segment_database(base_path, props)
```

```python
''' store / load segments dataset '''

uvsirlibs.database.storage.store_to_file(seg_db, '../data/', 'segments.db')

#seg_db = uvsirlibs.database.storage.load_from_file('../data/melodie-audio/', 'segments.db')
```

```python
''' select segments database subset '''

# filter segments by params
seg_db_ = [ node for node in seg_db ]#if
    #'_(Full)_' in node['track_name'] and
    #node['length'] >= 100
#]

print(len(seg_db_))
```

```python
''' listen to segments by subset index '''

seg_id = 8000
# get segment node
node = seg_db_[seg_id]

# define track import file path
file_path = '/'.join([node['track_path'], node['track_name']])

# import track data from file
_track = uvsirlibs.parse.parse_audio_file(file_path)

# unpack audio data
raw_audio = _track['raw_audio']
sample_rate = _track['sample_rate']


hop_length = 2048

# get start and end samples
start = node['start_frame'] * hop_length
end = node['end_frame'] * hop_length


# get raw audio data within segment range
seg_audio = raw_audio[:, start: end]


## segment audio playback
ipd.Audio(seg_audio, rate = sample_rate)
```

```python
''' feature extraction from segments subset '''

# define list features to extract
feat_list = []

# extract list of features from segments
feat_db = uvsirlibs.seg_extract_feats(seg_db_, feat_list)
```

```python
''' store / load features dataset '''

uvsirlibs.database.storage.store_to_file(feat_db, '../data/', 'features.db')

#feat_db = uvsirlibs.database.storage.load_from_file('../data/melodie-audio/', 'features.db')
```

```python
''' select features database subset '''

# filter features by params
feat_db_ = [ node for node in feat_db ]#if
              #(node['chroma-note_delta'] >= -5) and
            #(node['chroma-note_delta'] <= 5) and
            #(node['seg_track_length'] >= 0.01) and
            #(node['seg_track_length'] <= 0.2)
         #]

print(len(feat_db_))
```

```python
''' select feature for aggregation '''

feats = [
    #'track_name',
    #'track_path',
    
    #'seg_start_frame',
    #'seg_end_frame',
    #'seg_mid_frame',
    
    #'track_length',
    'track_tempo',
    
    'seg_length',
    'seg_track_length',
    'seg_track_depth',
    
    'energy-rms',
    #'energy-max',
    #'energy-min',
    'energy-var',
    
    'energy_h-rms',
    #'energy_h-max',
    #'energy_h-min',
    'energy_h-var',
    
    'energy_p-rms',
    #'energy_p-max',
    #'energy_p-min',
    'energy_p-var',
    
    'chroma-prime_note',
    'chroma-note_var',
    #'chroma-start_note',
    #'chroma-end_note',
    'chroma-note_delta'
]

# aggregate segment features, labels
agg = []
lbls = []

# iterate over segment feature subset
for node in feat_db_:
    
    # select feature values by name
    feat = [ node[f] for f in feats ]
    label = node['track_name'].split('.')[0]
    
    # store segment feature values, labels
    agg.append(feat)
    lbls.append(label)
    
# stack feature value dataset in np array
agg = np.stack(agg)

print(agg.shape)
```

```python
''' store dataset for classifier training '''

# pack dataset
to_class = {
    'features': feats,
    'data': agg,
    'labels': lbls,
}

uvsirlibs.database.storage.store_to_file(to_class, '../data/', 'train_dataset.db')

#dataset = uvsirlibs.database.storage.load_from_file('../data/melodie-audio/process/', 'dataset.db')
```
